```python
from dataloader import data_info, train_test_split_data
from classifiers import Classifier
from gain_lift import gain_lift_chart
```

```python
X_train, X_test, y_train, y_test = train_test_split_data(0.33)
model = Classifier()
```

### Gain and Lift charts for various models

Lift and Gain Charts are a useful way of visualizing how good a predictive model.They measure how much better one can expect to do with the predictive model comparing without a model. It can be helpful to find the best predictive model among multiple challenger models.

```python
""" KNeighbors Classifiers """
clf = model.KNeighbors(X_train, y_train)
gain_lift_chart(clf, X_test, y_test)
```

```python
""" Decision Tree Classifier """
clf = model.Decision_Tree(X_train, y_train)
gain_lift_chart(clf, X_test, y_test)
```

```python
""" Logistic Regression """
clf = model.Logistic_Reg(X_train, y_train)
gain_lift_chart(clf, X_test, y_test)
```

```python
""" Random Forest Classifier """
clf = model.Random_Forest(X_train, y_train)
gain_lift_chart(clf, X_test, y_test)
```

```python
""" Gaussian Naive Bayes Classifier"""
clf = model.Gaussian(X_train, y_train)
gain_lift_chart(clf, X_test, y_test)
```
